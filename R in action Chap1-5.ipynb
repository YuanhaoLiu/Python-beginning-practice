## Understanding dataset

Rectangular array of data with rows and columns -- Datasets  

Observations(rows) and variables(columns)/ records and fields/ examples and attributes

**Objects** -- anything that can be assigned to a variable, constants, data structures, functions, graphs. Have mode and class

**Data frame** -- hold data, main structure

**Factors** -- nominal or ordinal variables



### Data structures

Vector   一维      a <- c()  -- function c is used to store  *Cant mix modes in the same vector (numeric,character,or logical)*

Matrix   二维     each element same mode   myMatrix <- matrix()
 
Array    三维     myArray <- array(vector, dimensions, dimnames)

Dataframe  contain different modes of data     myData <- data.frame(col1, col2, col3,...)

List（includes vectors, arrays, dataframes, lists)

```r
# Example for matrix

y <- matrix (1:20, nrow=5, ncol=4)
y
```

```r
# byrow 意思就是是否按照一行行优先排列
# dimnames=list(char_vector_rownames, char_vector_colnames)

cells <- c(1,2,3,4)
rnames <- c("R1","R2")
cnames <- c("C1", "C2")
myMatrix <- matrix(cells, nrow=2, ncol=2, byrow=TRUE, dimnames=list(rnames,cnames))
myMatrix
```

```r
cells <- c(1,2,3,4)
rnames <- c("R1","R2")
cnames <- c("C1", "C2")
myMatrix <- matrix(cells, nrow=2, ncol=2, byrow=FALSE, dimnames=list(rnames,cnames))
myMatrix
```

```r
# About factor
# Categorical (nominal) and ordered (ordinal) variables in R are called factors.

diabetes <- c("Type1","Type2","Type1","Type1")
diabetes <- factor(diabetes)
```

```r
# assign the levels as 1=poor, 2=improved, 3=excellent

status <- factor(status, order=TRUE, levels=c("Poor","Improved","Excellent"))
```

```r
# Example Using factors

patientID <- c(1,2,3,4)
age <- c(20,30,40,50)
diabetes <- c("Type1","Type2","Type1","Type1")
status <- c("Poor","Improved","Excellent","Poor")
diabetes <- factor(diabetes)
status <-factor(status, order=TRUE)
patientData <- data.frame(patientID,age,diabetes,status)
patientData
str(patientData)
```

```r
manager <- c(1,2,3,4)
date <- c("10-1","10-2","10-3","10-4")
country <- c("US","US","UK","CN")
gender <- c("M","F","M","F")
age <- c(32,33,99,43)
q1 <- c(3,4,3,4)
q2 <- c(3,4,5,2)
q3 <- c(4,5,NA,5)
q4 <- c(2,4,5,1)
leader <- data.frame(manager, date, country, gender, age, q1, q2, q3, q4, stringAsFactors=FALSE) # 最后那个是说明只是plain texts
leader
```

```r
leader$age[leader$age == 99] <-NA  # variable[condition] <- expression will only make the assignment when condition is TRUE
leader
```
