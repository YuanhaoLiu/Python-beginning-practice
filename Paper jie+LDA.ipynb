```python
import gensim
from gensim import corpora, models, similarities
from gensim.models import LdaModel
from gensim.corpora import Dictionary
import jieba
import jieba.posseg as pseg
from string import punctuation
```

```python
jieba.load_userdict('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\userdict.txt')
stopwords = open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\stopwords.txt', 'r', encoding = 'utf8').readlines()
stopwords = [w.strip() for w in stopwords]
```

```python
add_punc = ', 。、【】 “” ：（）《》‘’{}？！%'
all_punc = punctuation + add_punc

tr = []  # 创建一个list

# 写入要打开的中文文件，注意encoding 
# 加载停用词并中文结巴

fr = open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\patent.txt','r',encoding='utf-8')
for w in fr.readlines():
    w = w.strip()      # 移除字符串首尾指定的字符
    w = "".join(w.split())
    if not len(w):                    # 看是否是空行
        continue
    outstr = ''                          # 给一个字符串
    seg_list = jieba.cut(w, cut_all=False)   # 精确结巴
    for word in seg_list:
        if word not in stopwords:            # 看是否在stopwords中
            if word != '\t':                 # ！= 不是制表符
                outstr += word
                outstr += " "
            if word in all_punc:
                outstr.remove(word)
    tr.append(outstr.strip().split(" "))    # 往这个list里取出空白且分割


# 写入新文件中    
# 注意这个tr的class 是个list

with open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\patentlist.txt','w',encoding = 'utf-8') as file:
    file.write(str(tr))
```

```python
with open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\patentlist.txt','r',encoding='utf-8') as file:
    texts = file.readlines()
keywords = jieba.analyse.extract_tags(str(texts), topK=50, withWeight=True, allowPOS=('nr','ns','nt','nz','n','vn','v'))
```

```python
dictionary = corpora. Dictionary(tr)
texts = [dictionary.doc2bow(text) for text in tr]

lda = LdaModel(corpus=texts, id2word=dictionary, num_topics=10)


lda.print_topics(10)
```
