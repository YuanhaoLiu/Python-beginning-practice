## JIEBA

```python
import jieba
import jieba.posseg as pseg
import re
```

```python
# load 各种user和stopwords

jieba.load_userdict('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\userdict.txt')
stopwords = open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\stopwords.txt', 'r', encoding = 'utf8').readlines()
stopwords = [w.strip() for w in stopwords]
```

```python
tr = []

fr = open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\paperdtm.txt','r',encoding='utf-8')
for w in fr.readlines():
    w = w.strip()      # 移除字符串首尾指定的字符
    w = "".join(w.split())
    if not len(w):                    # 看是否是空行
        continue
        
    outstr = ''                          # 给一个字符串
    
    w = re.sub(r'[A-Za-z0-9]|/d+','',str(w)) # 正则一下
    
    seg_list = jieba.cut(w, cut_all=False)   # 精确结巴
    for word in seg_list:
        if word not in stopwords:            # 看是否在stopwords中
            if word != '\t':                 # ！= 不是制表符
                outstr += word
                outstr += " "
            
    tr.append(outstr.strip().split(" "))    # 往这个list里取出空白且分割
```

```python
with open('C:\\Users\\mskim4\\Downloads\\.ipynb_checkpoints\\paperdtmlist.txt','w',encoding = 'utf-8') as file:
    file.write(str(tr))
```
