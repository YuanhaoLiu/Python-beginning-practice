### 实例：

```python
'''

Using ast.literal_eval(), as opposed to the dangerous eval() that you shouldn't use is safe, 
as it only evaluates literals that can have no side effects

'''

import ast

list2 = [inner for item in texts for inner in ast.literal_eval(item)] 
```

```python
# 其实list2 就可以用了， 但这个自定义可以把所有的全部展开为1个，感觉还挺有用的

def flatten(lst):
    for el in lst:
        if isinstance(el, list):
            yield from el
        else:
            yield el
```

```python
text = list(flatten(list2))
```

```python
from yellowbrick.text import DispersionPlot
```

```python
target_words = ['智慧城市','路灯','大数据','农业', '物流']
visualizer = DispersionPlot(target_words, color = 'k')
visualizer.fit(list2)
visualizer.show()
```

## 2. Token Frequency Distribution

```python
%matplotlib inline


from sklearn.feature_extraction.text import CountVectorizer

from yellowbrick.text import FreqDistVisualizer

vectorizer = CountVectorizer()
docs = vectorizer.fit_transform(texts)
features = vectorizer.get_feature_names()
```

```python
visualizer = FreqDistVisualizer(features=features, n=30, orient='v', size = (700,480), color = 'k')
visualizer.fit(docs)
plt.xticks(rotation=60)
plt.tick_params(labelsize = 16)
visualizer.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

from yellowbrick.text import TSNEVisualizer
from yellowbrick.datasets import load_hobbies

# Load the data and create document vectors

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(texts)
y = texts.target

# Create the visualizer and draw the vectors
tsne = TSNEVisualizer()
tsne.fit(X, y)
tsne.show()
```
